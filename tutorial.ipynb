```python
import pandas as pd

from src.utils.data_preparation import DataPrepare
from src.CatBoost.CB_inference import CatBoostCLF
from src.XGBoost.XGBoost_inference import XGBoostCLF

PATH_TO_DATA = './data/dataset_example.csv'
```

```python
df = pd.read_csv(PATH_TO_DATA)
df.head()
```

```python
prepare = DataPrepare(PATH_TO_DATA)
df_prepare = prepare.preprocess()
print(df_prepare.shape)
df_prepare.head()
```

```python
model_catboost = CatBoostCLF()
model_xgboost = XGBoostCLF()

predict_catboost = model_catboost.predict(df_prepare)
predict_xgboost = model_xgboost.predict(df_prepare)
predict_xgboost_with_threshold = model_xgboost.predict(df_prepare, threshold=True)
```

```python
df_prepare['predict_catboost'] = predict_catboost
df_prepare['predict_xgboost'] = predict_xgboost
df_prepare['predict_xgboost_with_threshold'] = predict_xgboost_with_threshold
```
